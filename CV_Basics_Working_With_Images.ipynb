## Working with Images using Scikit-image Library


### 1-Reading the data
Let's read the uploaded image file. We will use this file throughout this notebook. We will use "Scikit-image" library to read the image and make changes to it.

```
from matplotlib.pyplot import imread
 
image = imread('/content/drive/My Drive/Computer_vision/IMG_1248.jpg')
 
# Let's check the type of this image data
print("Type:", type(image))
 
# See the shape of it
print("Shape:", image.shape)
```

```
import numpy as np
import matplotlib.pyplot as plt
 
fig=plt.figure(figsize=(14, 28))
colors = {1: 'Red channel', 2: 'Green channel', 3:'Blue channel'}
for i in range(0, 3):
    # Get the corresponding color channel, last index is color index (channel)
    image_channel = image[:, :, i]
    fig.add_subplot(1, 3, i+1)
    # Plot the channel image and add title
    plt.imshow(image_channel, cmap='gray')
    plt.title(colors[i+1])
plt.show()
```

### 3-Resizing Images
We will use resize() function to fit the images to a desired size. Our original image has 325 rows and 579 columns. Let set the rows and columns to 120 and 250 respectively.

```
from skimage.transform import resize
 
image = imread('/content/drive/My Drive/Computer_vision/IMG_1248.jpg')
 
# Resized image
resized_im = resize(image, output_shape=(120, 250))
 
fig=plt.figure(figsize=(25, 80))
 
# Plot original image
fig.add_subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image with size: " + str(image.shape))
 
# Plot resized image
fig.add_subplot(1, 2, 2)
plt.imshow(resized_im)
plt.title("Resized Image with size: " + str(resized_im.shape))
 
plt.show()
```

## Working with Images using OpenCV

### 1-Reading the data
Let's read the uploaded image file. We will use this file throughout this notebook. We will use "OpenCV" library to read the image and make changes to it.

```
import cv2
 
image2 = imread('/content/drive/My Drive/Computer_vision/IMG_1248.jpg')
 
# Let's check the type of this image data
print("Type:", type(image2))
 
# See the shape of it
print("Shape:", image2.shape)
```

### 2-Showing the image
We will use the "matplotlib" library. Let's plot the image below. There are a few points here.
OpenCV store image channels in this order: Blue Green Red 
Without changing the order it will look like below


```
import matplotlib.pyplot as plt
 
plt.imshow(image2)
plt.show()
```

We can change the order to Red Green Blue using the cvtColor() function below.

```
image_color_order_changed = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image_color_order_changed)
plt.show()
```

Let's get three images for the three channels: Blue, Green and Red and plot them

```
import numpy as np
import matplotlib.pyplot as plt
 
fig=plt.figure(figsize=(14, 28))
color_channels = {1: 'Blue channel', 2: 'Green channel', 3:'Red channel'}
for i in range(0, 3):
    # Get the corresponding color channel, last index is color index (channel)
    image_channel = image2[:, :, i]
    fig.add_subplot(1, 3, i+1)
    # Plot the channel image and add title
    plt.imshow(image_channel, cmap='gray')
    plt.title(color_channels[i+1])
plt.show()
```

### 3-Resizing Images
We will use resize() function to fit the images to a desired size. Our original image has 794 rows and 1059 columns. Let set the rows and columns to 150 and 300 respectively.
We can try different resizing methods with OpenCV. We call those methods Interpolation Methods:
- Nearest neighbor interpolation (cv2.INTER_NEAREST): Applies nearest neighbor interpolation. In this method, the nearest pixel values are copied over. This usually results in not very pleasent results. The image mostly looks jagged.

- Linear interpolation (cv2.INTER_LINEAR): Applies linear interpolation. It fixes the jagged appereance we have after nearest neighbor interpolation by averaging pixel values. Gives smoother results than nearest neighbor method.

- Bicubic interpolation (cv2.INTER_CUBIC): Applies bicubic interpolation. It usually gives smoother results than linear interpolation.


```
import numpy as np
import matplotlib.pyplot as plt
 
# Let's use the RGB channel image
image_color_order_changed = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
 
fig=plt.figure(figsize=(20, 50))
 
nearest_neighbor_resized_im = cv2.resize(image_color_order_changed, dsize=(250, 120), interpolation=cv2.INTER_NEAREST)
linear_interp_resized_im = cv2.resize(image_color_order_changed, dsize=(250, 120), interpolation=cv2.INTER_LINEAR)
bicubic_interp_resized_im = cv2.resize(image_color_order_changed, dsize=(250, 120), interpolation=cv2.INTER_CUBIC)
 
fig.add_subplot(1, 3, 1)
plt.imshow(nearest_neighbor_resized_im)
plt.title('Nearest Neighbor method')
 
fig.add_subplot(1, 3, 2)
plt.imshow(linear_interp_resized_im)
plt.title('Linear Interpolation')
 
fig.add_subplot(1, 3, 3)
plt.imshow(bicubic_interp_resized_im)
plt.title('Bicubic Interpolation')
 
plt.show()
```
